```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
rawDF = pd.read_csv("mxmh.csv")

print("Created dataframe from csv data")
```

```python
cleanedDF = rawDF.drop('BPM',axis=1)

print("Removed the 'BPM' column from the dataset (107 na values)")
```

Q1: What was the most commonly used streaming service by participants?

```python
fig, ax = plt.subplots(1,figsize=(10,10))

streamGroup = cleanedDF.groupby("Primary streaming service").size()

plt.title("Primary streaming service")
ax.pie(x=streamGroup,
       labels=streamGroup.index,
       autopct = '%1.0f%%',
       colors=['pink','gray','orange','blue','green','red'])
```

```python
plt.title("Hours of music listened to daily")
plt.hist(cleanedDF['Hours per day'],color="green")
plt.xlabel("Hours")
plt.ylabel("Entries")

plt.show()
```

```python
# What is the most popular genre of music

fig, ax = plt.subplots(1,figsize=(10,10))

favGroup = cleanedDF.groupby("Fav genre").size()

plt.title("Primary streaming service")
ax.pie(x=favGroup,
       labels=favGroup.index,
       autopct = '%1.0f%%',
       colors=['cyan','magenta','yellow','green'])
```

```python
avgAge = round(cleanedDF['Age'].mean(),1)
avgHPD = round(cleanedDF['Hours per day'].mean(),1)
avgAnxiety = round(cleanedDF['Anxiety'].mean(),1)
avgDepression = round(cleanedDF['Depression'].mean(),1)
avgInsomnia = round(cleanedDF['Insomnia'].mean(),1)
avgOCD = round(cleanedDF['OCD'].mean(),1)

print("Average age of participants: " + str(avgAge))
print("Average listening hours of participants: " + str(avgHPD))
print("Average anxiety level of participants: " + str(avgAnxiety))
print("Average depression level of participants: " + str(avgDepression))
print("Average insomnia level of participants: " + str(avgInsomnia))
print("Average OCD level of participants: " + str(avgOCD))
```

```python
plt.title("Age range of participants")
plt.hist(cleanedDF['Age'],color="red")
plt.xlabel("Age")
plt.ylabel("# of participants")

plt.show()

plt.title("Anxiety level of participants")
plt.hist(cleanedDF['Anxiety'],color="orange")
plt.xlabel("Anxiety")
plt.ylabel("# of participants")

plt.show()

plt.title("Depression level of participants")
plt.hist(cleanedDF['Depression'],color="purple")
plt.xlabel("Depression")
plt.ylabel("# of participants")

plt.show()

plt.title("Insomnia level of participants")
plt.hist(cleanedDF['Insomnia'],color="green")
plt.xlabel("Insomnia")
plt.ylabel("# of participants")

plt.show()

plt.title("OCD level of participants")
plt.hist(cleanedDF['OCD'],color="blue")
plt.xlabel("OCD")
plt.ylabel("# of participants")

plt.show()

```

```python
plt.scatter(cleanedDF['Age'],
            cleanedDF['Anxiety'],
            c=["orange"],
            s=5
            )

plt.title("Correlation between age and anxiety level")
plt.xlabel("Age")
plt.ylabel("Anxiety level")

plt.show()

plt.scatter(cleanedDF['Age'],
            cleanedDF['Depression'],
            c=["purple"],
            s=5
            )

plt.title("Correlation between age and depression level")
plt.xlabel("Age")
plt.ylabel("Depression level")

plt.show()

plt.scatter(cleanedDF['Age'],
            cleanedDF['Insomnia'],
            c=["green"],
            s=5
            )

plt.title("Correlation between age and insomnia level")
plt.xlabel("Age")
plt.ylabel("Insomnia level")

plt.show()

plt.scatter(cleanedDF['Age'],
            cleanedDF['OCD'],
            c=["blue"],
            s=5
            )

plt.title("Correlation between age and OCD level")
plt.xlabel("Age")
plt.ylabel("OCD level")

plt.show()

```
